## Website scrapping : (Extract)

```python
class MangaDbSpider(scrapy.Spider):
    name = 'Manga'
    start_urls = ["https://myanimelist.net/anime.php",]

    def parse(self, response):
        xp = "//div[@id='horiznav_nav']//li/a/@href"
        return (Request(url, callback=self.parse_manga_list_page) for url in response.xpath(xp).extract())

    def parse_manga_list_page(self, response):
        for tr_sel in response.css('div.js-categories-seasonal tr ~ tr'):
            yield {
                "Title":  tr_sel.css('a[id] strong::text').extract_first().strip(),
                "Synopsis": tr_sel.css("div.pt4::text").extract_first(),
                "Format": tr_sel.css('td:nth-child(3)::text').extract_first().strip(),
                "Episodes": tr_sel.css('td:nth-child(4)::text').extract_first().strip(), 
                "Rating" : tr_sel.css('td:nth-child(5)::text').extract_first().strip()
                "Genre" : tr_sel.css("span.dark_text::text").extract_first().strip(),
                "Studio" : ,
                "Source" : ,
                "Duration" :
            
            }

            
        next_urls = response.xpath("//div[@class='spaceit']//a/@href").extract()
        for next_url in next_urls:
            yield Request(response.urljoin(next_url), callback=self.parse_anime_list_page)
```

#### (TEST GET IMAGE URL) With BeautifulSoup

```python
import requests 
from bs4 import BeautifulSoup 
import pandas as pd
import string
import json

def parse_episodes(listt):
    result = []
    for i in listt:
        r = i.strip()
        result.append(r)
    return result

page = 0
while page != 100:
    url = f"https://myanimelist.net/topanime.php?limit={page}"
    response = requests.get(url)
    doc = BeautifulSoup(response.text)
    
    row_content = doc.find_all('tr', {'class' : "ranking-list"})
    top_anime = []
    for row in row_content:
        episode = parse_episodes(row.find('div', class_ = "information di-ib mt4").text.strip().split('\n'))
        ranking = {
            'Title': row.find('div', class_="di-ib clearfix").find('a').text,
            'Image_URL': row.find('td', class_ ='title al va-t word-break').find('img')['data-src'],
            'Dates': episode[1]
        }
        top_anime.append(ranking)
    page = page + 1
        
jsonString = json.dumps(top_anime)
jsonFile = open("url_image_and_dates.json", "w")
jsonFile.write(jsonString)
print("done")
jsonFile.close()
```

```python
import pandas as pd 
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

```python
driver = webdriver.Chrome('C:/Users/Maathess/Desktop/chromedriver')
wait = WebDriverWait(driver,0)
```

```python
links = []
for page in range(0,1050,50) :
    driver.get("https://myanimelist.net/topanime.php?limit="+ str(page))
    link = driver.find_elements(By.CSS_SELECTOR, 'div[class="detail"] h3 a')
    for item in link :
        links.append(item.get_attribute('href'))
```

```python
titles = []
ratings = []
genres = []
studios = []
sources = []
status = []
types_arr = []
synopsis = []
episodes = []
dates = []
durations = []
images_url = []
Status = []



for item_link in links :
    driver.get(item_link)
    title = wait.until(EC.visibility_of_element_located((By.CSS_SELECTOR, 'div[class="h1-title"]')))
    rating = wait.until(EC.visibility_of_element_located((By.CSS_SELECTOR, 'div[class*="score-label"]')))
    genre = wait.until(EC.visibility_of_element_located((By.XPATH, '// *[contains(text(), "Genre")]/parent::div')))
    studio = wait.until(EC.visibility_of_element_located((By.XPATH, '// *[contains(text(), "Studio")]/parent::div')))
    source = wait.until(EC.visibility_of_element_located((By.XPATH, '// *[contains(text(), "Source")]/parent::div')))
    types = wait.until(EC.visibility_of_element_located((By.XPATH, '// *[contains(text(), "Type:")]/parent::div')))
    story = wait.until(EC.visibility_of_element_located((By.CSS_SELECTOR, 'p[itemprop="description"]')))
    episode = wait.until(EC.visibility_of_element_located((By.XPATH, '// *[contains(text(), "Episode")]/parent::div')))
    date = wait.until(EC.visibility_of_element_located((By.XPATH, '//span[contains(., "Aired")]/parent::div')))
    duration = wait.until(EC.visibility_of_element_located((By.XPATH, '// *[contains(text(), "Duration")]/parent::div')))
    image_url = wait.until(EC.visibility_of_element_located((By.CSS_SELECTOR, 'img[class=" lazyloaded"]')))
    statu = wait.until(EC.visibility_of_element_located((By.XPATH, '//span[contains(., "Status")]/parent::div')))
    
    titles.append(title.text)
    ratings.append(rating.text)
    genres.append(genre.text)
    studios.append(studio.text)
    sources.append(source.text)
    types_arr.append(types.text)
    synopsis.append(story.text)
    episodes.append(episode.text)
    dates.append(date.text)
    durations.append(duration.text)
    images_url.append(image_url.get_attribute('src'))
    status.append(statu.text)
```

```python
mydata = {
    'Title' : titles, 
    'Rating' : ratings, 
    'Genres' : genres,
    'Studio' : studios,
    'Sources' : sources,
    'Status' : status,
    'Type' : types_arr,
    'Synopsis' : synopsis,
    'Episodes' : episodes,
    'Date' : dates,
    'Duration' : durations,
    'Img_Url' : images_url,
    'Status' : status
    }

df = pd.DataFrame.from_dict(mydata, orient='index')
df = df.transpose()
df
```

```python
df['Genres'] = df['Genres'].str.replace('Genres:','')
df['Studio'] = df['Studio'].str.replace('Studios:','')
df['Studio'] = df['Studio'].str.replace('Producers:','')
df['Sources'] = df['Sources'].str.replace('Source:','')
df['Status'] = df['Status'].str.replace('Status:','')
df['Type'] = df['Type'].str.replace('Type:','')
df['Episodes'] = df['Episodes'].str.replace('Episodes:','')
df['Date'] = df['Date'].str.replace('Aired:','')
df['Duration'] = df['Duration'].str.replace('Duration:','')
df['Synopsis'] = df['Synopsis'].str.replace('\n\n','')
df
```

```python
df = df.drop('Genres', axis=1).join(df['Genres'].str.split(',', expand=True).stack().reset_index(level=1, drop=True).rename('Genres'))
df = df.drop('Studio', axis=1).join(df['Studio'].str.split(',', expand=True).stack().reset_index(level=1, drop=True).rename('Studio'))
```
